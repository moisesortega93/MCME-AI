## Cross validation procedure

Once we have defined the algorithm we can validate their performance.

In the Challenge 2 we defined the train, validation and test datasets. One of the habitual procedures to evaluate supervised methods the cross validation procedure:
* it creates automatically multiple train/validation partitions
* for each partition:
    * It trains a model with the train data
    * It uses the model trained to predict the samples in the validation partition
    * Error measures are computed with the predicted values and the ones in the dataset
* Error measures are averaged into mean errors and standard deviations




### Cross validation in surprise
Run 5-fold cross-validation (cv=5) and print results with surprise: <br>
https://surprise.readthedocs.io/en/stable/model_selection.html#cross-validation<br>
We have to provide the algorithm, the data that we use and the error measures that we want to get<br>
In this case we use Root Mean Squared Error (RMSE) and the Mean Absolute Error (MAE) to evaluate errors<br>
The cross validation averages multiple train/test datasets, we are not interested in particular predictions, but in the measures on the full dataset<br>

```python
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
