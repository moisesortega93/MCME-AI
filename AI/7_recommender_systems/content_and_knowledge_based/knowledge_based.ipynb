# Knowledge-based filtering

On this notebook, we will work with a movies dataset that can be downloaded from: https://www.kaggle.com/rounakbanik/the-movies-dataset (only movies_metadata.csv is required). <br> 

We will use this dataset to filter some movies according to some preferences of the user such as the revenue, the averaged vote and the length of the movie.

If there are some problems to download the file, this will be also provided at the classroom. Once downloaded, you need to upload it on the session storage (Files button on the menu from the left side).

```python
import pandas as pd
```

```python
#putting movies data on 'movies' dataframe
movies = pd.read_csv('movies_metadata.csv')
```

The knowlegde based filtering will filter characteristics of a database given the user selections. We will match this behaviour using the dataframe to filter a reduced number of films from the database

```python
movies.head(3)  
```

Lets filter by the revenue the movie had, we want to see a famous movie:

```python
filtered_movies = movies[movies['revenue']>100000000]
```

```python
len(filtered_movies)
```

We have too many films, lets continue filtering, now with the vote_average from the users:

```python
filtered_movies = filtered_movies[filtered_movies['vote_average']>8]
```

```python
len(filtered_movies)
```

Still we have more than 20 movies, now filter the duration of the movie, we want to spend about two hours seeing it:

```python
filtered_movies = filtered_movies[(110 < filtered_movies['runtime']) & (filtered_movies['runtime'] < 130)]
```

```python
len(filtered_movies)
```

We have less than 10 movies, lets sort them for the budget they had:

```python
filtered_movies.sort_values(by=['budget'],ascending=False)
```

Now we can chose between the remaining movies!
