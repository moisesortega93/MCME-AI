# Simple example with collaborative filtering

This notebook is based on the code from: https://realpython.com/build-recommendation-engine-collaborative-filtering/

```python
from scipy import spatial
```

In this example we have a set of users (from A to F). Each of them has evaluated 2 items (from 0 to 5)

```python
a = [1, 1.5]
b = [2, 3]
c = [2, 4]
d = [4, 2]
e = [0, 5]
f = [5, 0]
```

Lets put the evaluations in a structure for plotting the points in a graph:

```python
evaluations = {"a": a, "b": b, "c": c, "d": d, "e": e, "f":f}
print(evaluations)
```

```python
import matplotlib.pyplot as plt
for key in evaluations:
    plt.plot(evaluations[key][0],evaluations[key][1], 'o', color='black')
    plt.text(evaluations[key][0],evaluations[key][1], key, color='black')
```

## euclidean distance

```python
spatial.distance.euclidean(b, a)
```

```python
spatial.distance.euclidean(b, c)
```

```python
spatial.distance.euclidean(b, d)
```

```python
spatial.distance.euclidean(b, e)
```

```python
spatial.distance.euclidean(b, f)
```

In terms of distance, the ratings of C are the closest to those of B.

# cosine distance 

```python
spatial.distance.cosine(b,a)
```

```python
spatial.distance.cosine(b,c)
```

```python
spatial.distance.cosine(b,d)
```

```python
spatial.distance.cosine(b,e)
```

```python
spatial.distance.cosine(b,f)
```

In this case, user A and B have a 0 distance between them, making the most similar pair of users

Despite having different ratings, they have the same valoration of items, except for a factor. User B gives higher valorations than User A, but if we take out this factor, both users have the same preference for items

With this example, we see the benefit of using the cosine distance instead of the euclidean one
